```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model(
    model = ('groq:qwen/qwen3-32b')
)
# model.invoke("hello grok")
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title: str= Field(description="The movie title")
    year: int = Field(description="The year movie was released")
    director: str=Field(description="who directed this movie")
    Actor: str = Field(description="what was the name of this movie")
```

```python
model_with_structure=model.with_structured_output(Movie, include_raw=True)
```

```python
model_with_structure.invoke("give me details about Game of Throwns")
```

```python
# Type Dict

from typing_extensions import TypedDict,Annotated
class MovieDict(TypedDict):
    title: str= Annotated[str,"The movie title"]
    year: int = Annotated[int,"The year movie was released"]
    director: str=Annotated[str,"who directed this movie"]
    Actor: str = Annotated[str,"what was the name of this movie"]

model_with_structure=model.with_structured_output(Movie)
response=model_with_structure.invoke('Tell me about GOT')
```

```python

from langchain.agents import create_agent
from pydantic import BaseModel, Field



class Cricket(BaseModel):
    name_of_player : str= Field(description="name of player")
    dob : int = Field(description="year he was born")
    total_matches : int = Field(description="total matches played")
    total_sixes : int=Field(description="total 6's hit")

agent= create_agent(
    model = "groq:qwen/qwen3-32b",
    response_format = Cricket
)

result = agent.invoke({"messages":[{"role": "user", "content":"give me details about Yuvraj Singh"}]
})
result["structured_response"]
```
