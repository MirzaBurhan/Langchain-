```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model


model = init_chat_model(
    "llama-3.1-8b-instant",
    model_provider="groq",
)
```

```python
response=model.invoke("hello, how are you?")
response.content
```

```python
# Streaming and Batch
```

model.stream("write me a 200 word summary on AI")

```python
for chunk in model.stream("write me a 200 word paragraph on AI"):

    print(chunk.text, end ='', flush = True)
```

<!-- Bactching --- is the collection of independent req in the model can significantlly improve model, reduce cost, as the process can be done parallel.  -->

```python
response = model.batch([
    'why are parrots colorful?',
    'how does positional encoding work',
    'how to understand financial data'
])

for i in response:
    print(i)
```

```python
model.batch([
    'why are parrots colorful?',
    'how does positional encoding work',
    'how to understand financial data'
],
config={"max_concurrency": 5},
)
```
