```python
# Summarization


import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain.agents.middleware import SummarizationMiddleware
from langchain.messages import AIMessage, HumanMessage

@tool 
def get_weather(location: str)->str:
    """get the weather location"""
    return f"the weather for {location} is windy"

agent = create_agent(
    model = "groq:qwen/qwen3-32b",
    tools = [get_weather],
    middleware=[
        SummarizationMiddleware(
            model="groq:qwen/qwen3-32b",
            trigger=("tokens", 400),
            keep=("tokens", 200),
        ),
    ],
)

config = {"configurable" : {"thread_id": "test_1"}}


# counter token

def count_token(messages):
    total_chars = sum(len(str(m.content))for m in messages)
    return total_chars //4   



```

```python
# run_test
cities =["London","Paris","Tehran", "Delhi"]


for city in cities:
    response=agent.invoke({"messages":  [HumanMessage(content=f"find the hotel in {city}")]},
    config= config  




)


    tokens = count_token(response['messages'])
    print(f"{city}: ~{tokens} tokens, {len(response['messages'])} messages")
    print(f"{(response['messages'])}")



```
