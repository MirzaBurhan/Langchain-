```python
import os
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableSequence
from langchain.chat_models import init_chat_model
from langchain_core.output_parsers import StrOutputParser

from dotenv import load_dotenv
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
model = init_chat_model('groq:qwen/qwen3-32b')
```

```python
prompt_1= PromptTemplate.from_template(template='write a tweet on {topic}', input_variable=['topic'])
tweet_sequence = prompt_1 | model | StrOutputParser()
```

```python
prompt_2= PromptTemplate.from_template(template='write a post for Linkedin  {topic}', input_variable=['topic'])
linkedin_sequence = prompt_2 | model |StrOutputParser()
```

```python
from langchain_core.runnables import RunnableParallel
parallel = RunnableParallel(tweet_output = tweet_sequence, 
 linkedin_sequence= linkedin_sequence)
```

```python
response=parallel.invoke({"topic" : "AI"})
response
```
